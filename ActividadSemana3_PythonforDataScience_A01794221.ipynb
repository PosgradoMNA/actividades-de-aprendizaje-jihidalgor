```python
import time 
import sys
import numpy as np 

import matplotlib.pyplot as plt
%matplotlib inline  

def Plotvec2(a,b):
    ax = plt.axes()# to generate the full window axes
    ax.arrow(0, 0, *a, head_width=0.05, color ='r', head_length=0.1)#Add an arrow to the  a Axes with arrow head width 0.05, color red and arrow head length 0.1
    plt.text(*(a + 0.1), 'a')
    ax.arrow(0, 0, *b, head_width=0.05, color ='b', head_length=0.1)#Add an arrow to the  b Axes with arrow head width 0.05, color blue and arrow head length 0.1
    plt.text(*(b + 0.1), 'b')
    plt.ylim(-2, 2)#set the ylim to bottom(-2), top(2)
    plt.xlim(-2, 2)#set the xlim to left(-2), right(2)
```

```python
a = np.array([-1, 1])
b = np.array([ 1, 1])
Plotvec2(a,b)
print(np.dot(a,b))
```

```python
a = np.array([1, 0])
b = np.array([0, 1])
Plotvec2(a,b)
print(np.dot(a,b))
```

**REST APIs.**

Permiten comunicarse a través de internet permitiéndole aprovechar recursos como el almacenamiento, acceder a más datos algoritmos de inteligencia artificial, y mucho más.

Las siglas RE significan Representación, la sigla S significan Estado y la sigla T significan Transferencia.

En las REST APIs tu programa se llama cliente. La API se comunica con un servicio web al que llama a través de internet. Hay un conjunto de reglas relativas a la comunicación, la entrada o solicitud y la salida o respuesta.

```python
from nba_api.stats.static import teams
nba_teams = teams.get_teams()
nba_teams[:5]
```

```python
def one_dict(list_dict):
  keys = list_dict[0].keys()
  out_dict = {key:[] for key in keys}
  for dict_ in list_dict:
    for key, value in dict_.items():
      out_dict[key].append(value)
  return out_dict

dict_nba_team = one_dict(nba_teams)
```

```python
df_teams = pd.DataFrame(dict_nba_team)
df_teams.head()
```

```python
df_warriors = df_teams[df_teams['nickname']=='Warriors']
df_warriors
```

```python
id_warriors = df_warriors[['id']].values[0][0]
id_warriors
```

```python
from nba_api.stats.endpoints import leaguegamefinder
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable = id_warriors)
```

```python
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
games_home = games[games['MATCHUP']=='GSW vs. TOR']
games_away = games[games['MATCHUP']=='GSW @ TOR']
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
games_away.plot(x='GAME_DATE', y='PLUS_MINUS', ax=ax)
games_home.plot(x='GAME_DATE', y='PLUS_MINUS', ax=ax)
ax.legend(['away', 'home'])
plt.show()
```

**Aplicaction Program Interface (API).**

Transcribiremos un archivo de audio utilizando la API de texto a voz de Watson.
A continuación, traduciremos el texto a un nuevo idioma utilizando la API de traducción de idiomas de Watson.

En la llamada a la API, se enviará una copia del archivo de audio a la API; esto se denomina a veces una solicitud posterior. A continuación, la API enviará la transcripción del texto de lo que el individuo está diciendo. Por debajo, la API está haciendo una solicitud de obtención.

A continuación, enviamos el texto que queremos traducir a un segundo idioma a una segunda API. La API traducirá el texto y le enviará la traducción de vuelta; en este caso, nosotros traducimos del inglés al español.



*   Una clave API es una forma de acceder a la API. 
*   Es un conjunto único de caracteres que la API utiliza para identificación y autorización. Por lo general, la primera llamada a la API incluye la clave de la API. Esto permite acceder a la API.
*   Un endpoint es simplemente la ubicación del servicio. Se utiliza para encontrar la API en Internet, al igual que una dirección web.
