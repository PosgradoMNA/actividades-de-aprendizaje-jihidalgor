# Descriptive Statistics.

El análisis estadístico descriptivo ayuda a describir las características básicas de un conjunto de datos y obtener un breve resumen sobre la muestra y las medidas.

Una forma de hacerlo es utilizando la función la función describe() de pandas y aplicándola en el conjunto de datos. Esta función calcula automáticamente las estadísticas básicas para todas las variables numéricas. Muestra la media, el número total de datos, la desviación estándar, los cuartiles y los valores extremos. Los valores NaN se omiten
se omiten automáticamente en estas estadísticas. Esta función nos dará una idea más clara de la distribución que tienen las diferentes variables.

Una forma de resumir los datos categóricos es utilizando la función value_counts().

Los gráficos de caja-bigote son una buena forma de visualizar datos numéricos ya que se pueden visualizar las distintas distribuciones de los datos. Las principales características que muestra el gráfico de caja-bigote son la mediana de los datos, que representa dónde el punto medio de los datos. El cuartil superior muestra dónde está el percentil 75 y el cuartil inferior muestra dónde está el percentil 25. 
Los datos entre el Cuartil Superior y el Cuartil Inferior representan el Rango Intercuartil (IQR). A continuación, tenemos los extremos inferior y superior. Éstos se calculan como 1,5 veces el IQR por encima del percentil 75, y como 1,5 veces el IQR por debajo del percentil 25. Por último, los gráficos de caja también muestran los valores atípicos como puntos individuales que se encuentran fuera de los extremos superior
y los extremos inferiores. 

Una buena forma de visualizar variables continuas de nuestro conjunto de datos es utilizando un gráfico de dispersión (scatter plot). Cada observación en un gráfico de dispersión se representa como un punto. Este gráfico muestra la relación entre dos variables: 

* La variable predictora: es la variable que se utiliza para predecir un resultado.
* La variable objetivo: es la variable que se intenta predecir.

# GroupBy.

El método de "GroupBy" se utiliza en variables categóricas, agrupa los datos en subconjuntos según las diferentes categorías de esa variable.
Puede agrupar por una sola variable o puede agrupar por múltiples variables pasando en múltiples nombres de variables.

A pivot table has one variable displayed along the columns and the other variable displayed
along the rows.

Una "Pivot Table" tiene una variable mostrada a lo largo de las columnas y la otra variable mostrada
a lo largo de las filas.

Otra forma de representar la "Pivot Table" es mediante un gráfico de mapa de calor (heatmap). El mapa de calor toma una cuadrícula rectangular de datos y asigna una intensidad de color basada en el valor de los datos en los puntos de la cuadrícula. Es una gran manera de trazar la variable objetivo sobre múltiples variables y a través de esto obtener pistas visuales de la relación entre estas variables y el objetivo.

# Análisis de la varianza, ANOVA.

Para analizar variables categóricas podemos utilizar un método como
como el método ANOVA. Este métdod es una prueba estadística que significa "Análisis de la Varianza".

El ANOVA puede utilizarse para encontrar la correlación entre los diferentes grupos de una variable categórica.

La prueba ANOVA devuelve dos valores: la puntuación de la "F-score" y el "p-value". La "F-score" calcula la proporción de la variación entre la media de los grupos sobre la variación dentro de cada uno de los grupos de la muestra. El "p-value" muestra si el resultado obtenido es estadísticamente significativo. Por otra parte, la "F-test" calcula la relación de la variación entre medias de los grupos sobre la variación dentro de cada una de las medias de los grupos de la muestra.

La prueba ANOVA se puede realizar en Python utilizando el método f_oneway como la función incorporada del paquete Scipy.

```python
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```python
df_gptest
```

```python
grouped_test2.get_group('4wd')['price']
```

```python
from scipy import stats
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('rwd')['price'])  
   
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

# Correlación.

La correlación es una métrica estadística que permite medir hasta qué punto diferentes variables son interdependientes.
En otras palabras, cuando observamos dos variables a lo largo del tiempo, si una de ellas cambia, ¿Cómo afectó este cambio en la otra variable?

Es importante saber que la correlación no implica causalidad.

Una forma de medir la fuerza de la correlación entre variables numéricas continuas es utilizando un método llamado Correlación de Pearson. 

Este método entrega dos valores, el coeficiente de correlación y el valor p.

Para el coeficiente de correlación, un valor cercano a 1 implica una gran correlación positiva, mientras que un valor cercano a -1 implica una gran correlación negativa y un valor cercano a 0 implica que no hay correlación entre las variables.

A continuación, el valor p nos indicará el grado de certeza de la correlación que hemos calculado. Para el valor p, un valor inferior a 0,001 nos da una gran certeza sobre el coeficiente de correlación
que hemos calculado. Un valor entre 0,001 y 0,05 nos da una certeza moderada, un valor entre 0,05 y 0,1 nos nos da una certeza débil y un valor p mayor que 0,1 nos dará ninguna certeza de la correlación
en absoluto.

Podemos decir que existe una correlación fuerte cuando el coeficiente de correlación es cercano a 1 o -1 y el valor p es inferior a 0,001.
