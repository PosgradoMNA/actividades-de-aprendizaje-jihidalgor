# Ejercicio 1, Descomposición y Reconstrucción
***Descomposición***

Encuentra los eigenvalores y eigenvectores de las siguientes matrices

$A1 = \begin{pmatrix} 3,0,2 \\ 3,0,-2 \\ 0,1,1 \end{pmatrix}$
$A2 = \begin{pmatrix} 1,3,8 \\ 2,0,0 \\ 0,0,1 \end{pmatrix}$
$A3 = \begin{pmatrix} 5,4,0 \\ 1,0,1 \\ 10,7,1 \end{pmatrix}$

y reconstruye la matriz original a traves de las matrices $WDW^{-1}$ 

(OJO. Esto es lo mismo de la ecuación del paso 2 solo le cambiamos la variable a la matriz diagonal)


```python
import numpy as np
from numpy import array
from numpy.linalg import eig
```

Primer paso, definir la matriz.

$A1 = \begin{pmatrix} 3,0,2 \\ 3,0,-2 \\ 0,1,1 \end{pmatrix}$

```python
A1 = array([[3,0,2],[3,0,-2], [0,1,1]])
print(A1)
```

Segundo paso, calcular la eigendescomposición.

```python
values, vectors = eig(A1)
# D
print('---eigenvalores---')
print(values,'\n') 
# W
print('---eigenvectores---')
print(vectors)
```

Tercer paso, proceso de reconstrucción.

$C = W\,D\,W^{-1}$

```python
values, vectors = np.linalg.eig(A1)

W = vectors
Winv = np.linalg.inv(W)
D = np.diag(values)

B = np.dot(np.dot(W, D), Winv)
print(B.real.round(1))
```

Primer paso, definir la matriz.

$A2 = \begin{pmatrix} 1,3,8 \\ 2,0,0 \\ 0,0,1 \end{pmatrix}$

```python
A2 = array([[1,3,8],[2,0,0], [0,0,1]])
print(A2)
```

Segundo paso, calcular la eigendescomposición.

```python
values, vectors = eig(A1)
# D
print('---eigenvalores---')
print(values,'\n') 
# W
print('---eigenvectores---')
print(vectors)
```

Tercer paso, proceso de reconstrucción.

$C = W\,D\,W^{-1}$

```python
values, vectors = np.linalg.eig(A2)
W = vectors
Winv = np.linalg.inv(W)
D = np.diag(values)

B = np.dot(np.dot(W, D), Winv)
print(B.round(1))
```

Primer paso, definir la matriz.

$A3 = \begin{pmatrix} 5,4,0 \\ 1,0,1 \\ 10,7,1 \end{pmatrix}$

```python
A3 = array([[5,4,0],[1,0,1], [0,1,1]])
print(A3)
```

Segundo paso, calcular la eigendescomposición.

```python
values, vectors = eig(A1)
# D
print('---eigenvalores---')
print(values,'\n') 
# W
print('---eigenvectores---')
print(vectors)
```

Tercer paso, proceso de reconstrucción.

$C = W\,D\,W^{-1}$

```python
values, vectors = np.linalg.eig(A3)
W = vectors
Winv = np.linalg.inv(W)
D = np.diag(values)

B = np.dot(np.dot(W, D), Winv)
print(B.round(1))
```

De los resultados anteriores, podemos comprobar que efectivamente al contar con los eigenvalores y sus correspondientes eigenvectores es posible reconstruir la matriz original al realizar la operación:

$C = W\,D\,W^{-1}$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
plt.style.use('ggplot')

# Cargando los datos.
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Preprocesamiento de los datos con la función StandardScaler (Estandarización).
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

# Modelo de PCA.
pca = PCA(n_components=2)     # Se definen el número de componentes.
X_new = pca.fit_transform(X)  # Proyección de los datos originales en el espacio de PCA.
```

Visualización de los datos antes y después de PCA.

```python
fig, axes = plt.subplots(1,2)
axes[0].scatter(X[:,0], X[:,1], c=y)
axes[0].set_xlabel('x1')
axes[0].set_ylabel('x2')
axes[0].set_title('Before PCA')
axes[1].scatter(X_new[:,0], X_new[:,1], c=y)
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].set_title('After PCA')
plt.show()
```

Del espacio PCA se puede observar que la varianza se maximiza a lo largo de PC1 (explica el 73% de la varianza) y PC2 (explica el 22% de la varianza). 

Juntos, explican el 95%.

```python
print(pca.explained_variance_ratio_)
```

La prueba de varianza máxima se puede observar al calcular la matriz de covarianza del espacio reducido. 

Para ello, podemos emplear la función **np.conv()** de Numpy, 

```python
np.cov(X_new.T)
```

El resultado es un arreglo en cuya diagonal principal se tienen los valores de la varianza y son iguales a los eigenvalores actuales de la covariaza almacenados en:

*pca.explained_variance_*


```python
pca.explained_variance_
```

La importancia de cada característica se refleja en la magnitud de los valores correspondientes en los eigenvectores (**a mayor magnitud, mayor importancia**).

Para conocer los valores de las componentes principales, podemos utilizar la función:

*pca.components_*

*Nota: evaluar con su valor absoluto.*



```python
print(abs( pca.components_ ))
```

El resultado es un arreglo de 2 filas y 4 columnas.

Es importante recordar que pca.components tiene al forma [n_componentes, n_características]. Por lo tanto, se observa que la fila 1 corresponde al primer componente principal (PC1) y las características más importantes son la 1, 3 y 4 (columnas).

Por otra parte, se observa que la fila 2 corresponde al segundo componente principal (PC2) y las características más importantes son las columnas 1 y 2.

*En conclusión, observar los valores absolutos de los componentes de los eigenvectores correspondientes a los k mayores eigenvalores.* 

*En sklearn los componentes se ordenan por la varianza explicada. Cuanto más grandes son estos valores absolutos, más contribuye una característica específica a ese componente principal.* 

La mejor manera de visualizar el método de PCA es mediante un **biplot**.

```python
def biplot(score, coeff , y):
    '''
    Author: Serafeim Loukas, serafeim.loukas@epfl.ch
    Inputs:
       score: the projected data
       coeff: the eigenvectors (PCs)
       y: the class labels
   '''
    xs = score[:,0] # projection on PC1
    ys = score[:,1] # projection on PC2
    n = coeff.shape[0] # number of variables
    plt.figure(figsize=(10,8), dpi=100)
    classes = np.unique(y)
    colors = ['g','r','y']
    markers=['o','^','x']
    for s,l in enumerate(classes):
        plt.scatter(xs[y==l],ys[y==l], c = colors[s], marker=markers[s]) # color based on group
    for i in range(n):
        #plot as arrows the variable scores (each variable has a score for PC1 and one for PC2)
        plt.arrow(0, 0, coeff[i,0], coeff[i,1], color = 'k', alpha = 0.9,linestyle = '-',linewidth = 1.5, overhang=0.2)
        plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'k', ha = 'center', va = 'center',fontsize=10)

    plt.xlabel("PC{}".format(1), size=14)
    plt.ylabel("PC{}".format(2), size=14)
    limx= int(xs.max()) + 1
    limy= int(ys.max()) + 1
    plt.xlim([-limx,limx])
    plt.ylim([-limy,limy])
    plt.grid()
    plt.tick_params(axis='both', which='both', labelsize=14)
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault) # reset ggplot style
# Call the biplot function for only the first 2 PCs
biplot(X_new[:,0:2], np.transpose(pca.components_[0:2, :]), y)
plt.show()
```

Ahora se puede verificar visualmente lo siguiente:

* La varianza se maximiza.
* Las características 1, 3 y 4 son las más importantes para la componente principal 1 (PC1).
* Las características 1 y 2 son las más importantes para la componente principal 2 (PC2).


Por otra parte, las flechas (variables/características) que apuntan hacia la misma dirección indican correlación entre las variables que representan, mientras que las flechas que apuntan en dirección contraria indican un contraste entre las variables que representan.
